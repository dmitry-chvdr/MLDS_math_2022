## Задание 1

Найдите экстремумы следующих функций (вручную, где возможно) и укажите их тип (максимум, минимум):

a) $f(x) = x^2 - 30x + 1$

b) $g(x) = -x^3 + 5x - 7x^2 + 17$

c) $h(x) = x^2 - 1/x$

d) $j(x) = 1/x^5 + x^{17} + 8$

## Задание 2

Найдите экстремумы этих функций в Python.

```python
import sympy as sp
from sympy.abc import x
from sympy import S
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
func_a = x**2 - 30*x + 1
func_b = -x**3 + 5*x - 7*x**2 + 17
func_c = x**2 - 1/x
func_d = 1/x**5 + x**17 + 8

functions_to_solve = [func_a, func_b, func_c, func_d]
```

```python
def extremums(function_to_solve):
    function_diff = sp.diff(function_to_solve)
    function_diff_zeros = sp.solveset(function_diff, x, S.Reals)
    return tuple((zero, function_to_solve.subs(x, zero)) for zero in function_diff_zeros)

for func_to_solve in functions_to_solve:
    print(f"Экстремум функции {func_to_solve} и точка в которой он достигается")
    print(extremums(func_to_solve), end="\n\n")
```

```python
def inflection_points(function_to_solve):
    function_diff2 = sp.diff(function_to_solve, x, 2)
    function_diff2_zeros = sp.solveset(function_diff2, x, S.Reals)
    return tuple((zero, function_to_solve.subs(x, zero)) for zero in function_diff2_zeros)

for func_to_solve in functions_to_solve:
    print(f"Точки перегиба функции {func_to_solve}:")
    print(inflection_points(func_to_solve), end="\n\n")
```

## Задание 6

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(function_min, start, learning_rate, epsilon, delta, num_iterations, limit_iterations=None):
    x_curr = start
    df = sp.diff(function_min)

    trace = []
    trace.append(x_curr)

    for i in range(limit_iterations or num_iterations):
        df_value = df.subs(x, x_curr)
        x_new = x_curr - learning_rate * df_value
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_curr, trace

        if abs(df_value) < delta:
            return x_curr, trace

        x_curr = x_new

    return x_curr, trace
```

## Задание 8

Реализуйте [метод Ньютона](https://translated.turbopages.org/proxy_u/en-ru.ru.c97ef85d-631c4b6a-ff938028-74722d776562/https/en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def newton_method(function_min, start_a, start_b, num_iterations, limit_iterations=None):
    
    diff = sp.diff(function_min, x)
    diff2 = sp.diff(function_min, x, 2)
    diff3 = sp.diff(function_min, x, 3)
    
    x_start = start_a if diff.subs(x, start_a) * diff3.subs(x, start_a) > 0 else start_b
    x_curr = x_start

    for i in range(limit_iterations or num_iterations):
        h = diff.subs(x, x_curr) / diff2.subs(x, x_curr)
        x_curr = x_curr - h

    return x_curr, x_start
```
